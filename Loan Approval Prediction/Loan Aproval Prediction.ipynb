```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
loan_data_Approved = pd.read_csv("loan_data.csv")
```

```python
# replacing some categorical data with numerical number for easy evaluation
loan_data_Approved['Gender'] = loan_data_Approved['Gender'].map({'Male': 1, 'Female': 0})
loan_data_Approved['Married'] = loan_data_Approved['Married'].map({'Yes': 1, 'No': 0})
loan_data_Approved['Self_Employed'] = loan_data_Approved['Self_Employed'].map({'Yes': 1, 'No': 0})
loan_data_Approved['Education'] = loan_data_Approved['Education'].map({'Graduate': 1, 'Not Graduate': 0})
loan_data_Approved['Property_Area'] = loan_data_Approved['Property_Area'].map({'Urban': 2, 'Semiurban': 1, 'Rural':0})
loan_data_Approved['Loan_Status'] = loan_data_Approved['Loan_Status'].map({'Y': 1, 'N': 0})
```

```python
loan_data_Approved.head()
```

```python
loan_data_Approved.shape
```

```python
loan_data_Approved.isnull().sum()
```

```python
loan_data_Approved['Self_Employed'] = loan_data_Approved['Self_Employed'].fillna(loan_data_Approved['Self_Employed'].mode()[0])
loan_data_Approved['Credit_History'] = loan_data_Approved['Credit_History'].fillna(loan_data_Approved['Credit_History'].mode()[0])
loan_data_Approved['Gender'] = loan_data_Approved['Gender'].fillna(loan_data_Approved['Gender'].mode()[0])
loan_data_Approved['Dependents'] = loan_data_Approved['Dependents'].fillna(loan_data_Approved['Dependents'].mode()[0])
```

```python
loan_data_Approved['Loan_Amount_Term'] = loan_data_Approved['Loan_Amount_Term'].fillna(loan_data_Approved['Loan_Amount_Term'].median())
```

```python
loan_data_Approved.isnull().sum()
```

```python
loan_data_Approved.head()
```

```python
# converting the infinte number to integer
loan_data_Approved['Gender'] = loan_data_Approved['Gender'].fillna(0).replace([float('inf'), float('-inf')], 0).astype(int)
loan_data_Approved['Self_Employed'] = loan_data_Approved['Self_Employed'].fillna(0).replace([float('inf'), float('-inf')], 0).astype(int)
loan_data_Approved['Credit_History'] = loan_data_Approved['Credit_History'].fillna(0).replace([float('inf'), float('-inf')], 0).astype(int)

```

```python
loan_data_Approved.head()
```

```python
loan_data_Approved.isnull().sum()
```

```python
loan_data_Approved.shape
```

```python
loan_data_Approved["Loan_Status"].value_counts()
```

```python
# Replace '3+' with 3 and convert to numeric
loan_data_Approved['Dependents'] = loan_data_Approved['Dependents'].replace('3+', 3).astype(int)
```

```python
X_dropid = loan_data_Approved.drop("Loan_ID", axis =1)
```

```python
X = X_dropid.drop("Loan_Status", axis = 1)
Y = loan_data_Approved["Loan_Status"]
```

```python
X.head()
```

```python
Y.head()
```

```python
# converting the infinte number to integer
loan_data_Approved['CoapplicantIncome'] = loan_data_Approved['CoapplicantIncome'].fillna(0).replace([float('inf'), float('-inf')], 0).astype(int)
loan_data_Approved['LoanAmount'] = loan_data_Approved['LoanAmount'].fillna(0).replace([float('inf'), float('-inf')], 0).astype(int)
loan_data_Approved['Loan_Amount_Term'] = loan_data_Approved['Loan_Amount_Term'].fillna(0).replace([float('inf'), float('-inf')], 0).astype(int)
```

```python
loan_data_Approved.tail()
```

```python
# Spliting in to train and test 

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.1, stratify=Y, random_state=2) 
```

```python
print(X_test.shape, X_train.shape, X.shape)
```

```python
X_train.tail()
```

```python
scale_check_train =  np.std(X_train, axis=0)
scale_check_test = np.std(X_test, axis=0)
```

```python
# Fit the scaler on the training data and transform both training and testing sets
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print(X_train_scaled )
```

```python
scale_check_train_1 =  np.std(X_train_scaled, axis=0)
scale_check_test_1 = np.std(X_test_scaled, axis=0)
```

```python
print(scale_check_test_1)
```

Model Evaluation



```python
model=SVC(kernel='linear') 

model.fit(X_train_scaled, Y_train)
```

```python
X_train_prediction = model.predict(X_train_scaled)
training_data_accuracy =accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test_scaled)
test_data_accuracy =accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data: ', test_data_accuracy)
```
